```python
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
df = pd.read_csv('./housing.csv')
df
```

```python
x = df.drop('ocean_proximity', axis=1)
x = x.drop('total_bedrooms', axis=1)
x = x.drop('median_house_value', axis=1)
x
```

```python
y = df['median_house_value']
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=10)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
y_predict
```

```python
import matplotlib.pyplot as plt
```

```python
y_changed = pd.DataFrame({'Actual': y_test, 'Predicted': y_predict, 'Difference': y_test-y_predict})
y_changed
```

```python
plt.scatter(x=y_test, y=y_predict)
plt.xlabel("Original Values")
plt.ylabel("Predicted Values")
plt.legend()
```

Using normalization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
x_norm = scaler.fit_transform(x_train)
x_test_norm = scaler.fit_transform(x_test)
```

```python
model.fit(x_norm, y_train)
y_pred_norm = model.predict(x_test_norm)
```

```python
y_changed_norm = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred_norm, 'Difference': y_test-y_pred_norm})
y_changed_norm
```

```python
plt.scatter(x=y_test, y=y_pred_norm)
plt.xlabel("Original Values")
plt.ylabel("Predicted Values")
plt.legend()
```
